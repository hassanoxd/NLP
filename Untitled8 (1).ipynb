```
def nbrstopwords(doc):
  text = webtext.raw(doc)
  text_tokens = word_tokenize(text)
  #print('nombre de word \n',len(text_tokens))
  #print(text_tokens)
  count = 0
  stop_words = set(stopwords.words('english')) 
  for word in text_tokens:
     if word in stop_words:
       count = count + 1
  #print('nombre de stop words')
  return count

```

**eliminer les stopwords**

```
def eliminerstw(doc):
  text = webtext.raw(doc)
  text_tokens = word_tokenize(text)
  #print('nombre de word \n',len(text_tokens))
  #print(text_tokens)
  stop_words = set(stopwords.words('english')) 
  tokens_without_sw = [word for word in text_tokens if not word in stopwords.words()]
  return tokens_without_sw
```

```
eliminerstw('wine.txt')
```


**transformer toutes les majuscules en minuscules.**



```
def tranf(doc):
  text = webtext.raw(doc)
  text_tokens = word_tokenize(text)
  #print('nombre de word \n',len(text_tokens))
  #print(text_tokens)
  text_l= ' '.join([word.lower() for word in text_tokens])
  return text_l
```

```
tranf('wine.txt')
```

**les 10 tokens les plus fréquents**

```
from collections import Counter
def tok10(doc):
  text = webtext.raw(doc)
  text_tokens = word_tokenize(text)
  return Counter(webtext.raw(doc)).most_common(10)
```

```
tok10('firefox.txt')
```

**les 10 tokens les plus rares **

```
from collections import Counter
def tok_10(doc):
  text = webtext.raw(doc)
  text_tokens = word_tokenize(text)
  return Counter(webtext.raw(doc)).most_common()
```

```
tok_10('firefox.txt')
```

**élimine les chiffres et la ponctuation**

```
import re
def subn(doc):
  text = webtext.raw(doc)
  text_tokens = word_tokenize(text)
  punctuation = re.compile(r'[-.?!/\,":;()|0-9]')
  res = re.sub(punctuation, '', text) 
  return res
```

```
subn('firefox.txt')
```

```
def digittoword(doc):
  text = webtext.raw(doc)
  text_tokens = word_tokenize(text)
  result = ' '.join(word for word in text_tokens if not word.isdigit())
  return result
```

```
digittoword('firefox.txt')
```

```
 import inflect
 def toword(doc):
  text = webtext.raw(doc)
  text_tokens = word_tokenize(text)
  inflector = inflect.engine()
 
  words = inflector.number_to_words(text_tokens)
  return words
```

```
from nltk.stem import PorterStemmer
def stremmer(doc):
  text = webtext.raw(doc)
  text_tokens = word_tokenize(text)
  ps = PorterStemmer()
  stems=[ps.stem(word) for word in text_tokens]
  return stems
```

```
stremmer('firefox.txt')
```

```
from nltk.stem.lancaster import LancasterStemmer
def stremmerlan(doc):
  text = webtext.raw(doc)
  text_tokens = word_tokenize(text)
  ps = LancasterStemmer()
  stems=[ps.stem(word) for word in text_tokens]
  return stems
```

```
stremmerlan('firefox.txt')
```

```
from nltk.stem import WordNetLemmatizer 
import nltk
nltk.download('wordnet')
  
def lem(doc):
  text = webtext.raw(doc)
  text_tokens = word_tokenize(text)
  lemmatizer = WordNetLemmatizer()
  stems=[lemmatizer.lemmatize(word) for word in text_tokens]
  return stems
```

```
lem('firefox.txt')
```
